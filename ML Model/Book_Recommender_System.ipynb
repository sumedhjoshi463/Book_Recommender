# Popularity based Recommender System

we will show top 50 books with highest ratings which are rated by more than 250 users at leasat. 
In order to do that, we will merge ratings and books dataframes.


```python
ratings_with_book_name = ratings.merge(books, on='ISBN')
```

```python
num_rating_df=ratings_with_book_name.groupby("Book-Title").count()['Book-Rating'].reset_index()
num_rating_df.rename(columns = {'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df=ratings_with_book_name.groupby("Book-Title").mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns = {'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popular_books_df= num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_books_df
```

```python
popular_books_df[popular_books_df['num_ratings']>=250].sort_values('avg_ratings',ascending = False).head(50)
```

Now we want authors name, image url and publishers name if possible.so 
we merge popular_books_df and boos on book-title

```python
popular_books_df= popular_books_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

# Collaborative Filtering Based Rcommander System

```python
x = ratings_with_book_name.groupby('User-ID').count()['Book-Rating']>200
users_who_have_rated= x[x].index
```

```python
users_who_have_rated
```

```python
filtered_rating=ratings_with_book_name[ratings_with_book_name['User-ID'].isin(users_who_have_rated)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score= cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('The Great Gatsby')
```
